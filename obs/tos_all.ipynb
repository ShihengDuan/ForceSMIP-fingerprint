```python
def get_detection_obs(metrics):
    upper = np.array(metrics['s_obs_list'])+np.array(metrics['obs_se'])
    lower = np.array(metrics['s_obs_list'])-np.array(metrics['obs_se'])
    center = np.array(metrics['s_obs_list'])
    upper = upper/np.array(metrics['n_list'])
    lower = lower/np.array(metrics['n_list'])
    center = center/np.array(metrics['n_list'])
    upper = xr.DataArray(upper, dims=['time'], coords={'time': time})
    lower = xr.DataArray(lower, dims=['time'], coords={'time': time})
    center = xr.DataArray(center, dims=['time'], coords={'time': time})

    upper = upper.reindex(time=list(reversed(upper.time)))
    lower = lower.reindex(time=list(reversed(lower.time)))
    center = center.reindex(time=list(reversed(center.time)))

    indices = (upper < 2)
    first_indices = (indices.argmax(dim='time'))
    whole_period = upper.time.shape[0]
    back = whole_period-first_indices
    if back<whole_period:
        detect_time_upper = back
    else:
        detect_time_upper = None
    
    indices = (lower < 2)
    first_indices = (indices.argmax(dim='time'))
    whole_period = lower.time.shape[0]
    back = whole_period-first_indices
    if back<whole_period:
        detect_time_lower = back
    else:
        detect_time_lower = None

    indices = (center < 2)
    first_indices = (indices.argmax(dim='time'))
    whole_period = center.time.shape[0]
    back = whole_period-first_indices
    if back<whole_period:
        detect_time_center = back
    else:
        detect_time_center = None
    return detect_time_lower, detect_time_center, detect_time_upper
```

```python
def plot_detection_ax(ax, detect, cobe, pcmdi, ersst, cmip_label=False):
    for i, back in enumerate(detect):
        # plt.hist(back, alpha=0.2, density=True, color=colors[i])
        mu, std = norm.fit(back)
        x = np.linspace(xmin, xmax, 100)
        p = norm.pdf(x, mu, std)
        if cmip_label:
            ax.plot(x, p, linewidth=1, color=colors[i], label=models[i])
        else:
            ax.plot(x, p, linewidth=1, color=colors[i])

    detect_time_lower, detect_time_center, detect_time_upper = get_detection_obs(metrics=cobe)
    if detect_time_lower is None:
        detect_time_lower = xmax
        uncertain = None
    else:
        uncertain = detect_time_lower.data-detect_time_upper.data
    if uncertain is None:
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=cobe_color)
    else:
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=cobe_color, label='COBE '+str(round(uncertain, 2)))
    if cmip_label:
        ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=cobe_color, linewidth=2, label='COBE')
    else:
        ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=cobe_color, linewidth=2)
    
    detect_time_lower, detect_time_center, detect_time_upper = get_detection_obs(metrics=pcmdi)
    if detect_time_lower is None:
        detect_time_lower = xmax
        uncertain = None
    else:
        uncertain = detect_time_lower.data-detect_time_upper.data
    if uncertain is None:
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=pcmdi_color)
    else:
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=pcmdi_color, label='PCMDI '+str(round(uncertain, 2)))
    if cmip_label:
        ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=pcmdi_color, linewidth=2, label='PCMDI')
    else:
        ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=pcmdi_color, linewidth=2)
    
    detect_time_lower, detect_time_center, detect_time_upper = get_detection_obs(metrics=ersst)
    if detect_time_lower is None:
        detect_time_lower = xmax
        uncertain = None
    else:
        uncertain = detect_time_lower.data-detect_time_upper.data
    if uncertain is None:
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=ersst_color)
    else:
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=ersst_color, label='ERSST '+str(round(uncertain, 2)))
    if cmip_label:
        ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=ersst_color, linewidth=2, label='ERSST')
    else:
        ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=ersst_color, linewidth=2)
    
    
    if cmip_label:
        hs, ls = ax.get_legend_handles_labels()
        print(ls)
        hs = [hs[-6], hs[-4], hs[-2]]
        ls = [ls[-6], ls[-4], ls[-2]]
        ax.legend(hs, ls, loc='upper right')
    else:
        ax.legend(loc='upper right')
```

```python
def ax_plot_sn(ax, cmip, cobe, cmip_label=False):
    for i in range(5):
        data = cmip['model_results'][i]
        data = data['sn'].isel(time=-1)
        mu, std = norm.fit(data)
        x = np.linspace(xmin, xmax, 100)
        p = norm.pdf(x, mu, std)
        if cmip_label:
            ax.plot(x, p, linewidth=1, color=colors[i], label=models[i])
        else:
            ax.plot(x, p, linewidth=1, color=colors[i])
    
    s_obs_list = cobe['s_obs_list']
    s_se = cobe['obs_se']
    n_list = cobe['n_list']
    center = s_obs_list[-1]/n_list[-1]
    upper = (s_obs_list[-1]+s_se[-1])/n_list[-1]
    lower = (s_obs_list[-1]-s_se[-1])/n_list[-1]
    uncertain = upper.data-lower.data
    if cmip_label:
        ax.axvline(x=center, ymin=0, ymax=1, color=cobe_color, linewidth=2, label='COBE')
    else:
        ax.axvline(x=center, ymin=0, ymax=1, color=cobe_color, linewidth=2)
    ax.axvspan(upper, lower, alpha=0.2, color=cobe_color, label='COBE '+str(round(uncertain, 2)))
    ax.axvline(x=2, ymin=0, ymax=1, color='black', linewidth=1, linestyle='--')
    print(center)
    
    if cmip_label:
        hs, ls = ax.get_legend_handles_labels()
        hs = [hs[-1]]
        ls = [ls[-1]]
        ax.legend(hs, ls, loc='upper right')
    else:
        ax.legend(loc='upper right')
```

```python
colors = sns.color_palette('deep', 7)
cobe_color = colors[-2]
gpcp_color = colors[-1]
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 6), sharex=True, sharey=True)
xmin = -2
xmax = 50
ax = axes.flatten()[0]
ax.set_title('Anomaly', fontsize=12)
ax_plot_sn(ax, cmip=cmip, cobe=metrics_cobe)

ax = axes.flatten()[1]
ax.set_title('StdAnomaly', fontsize=12)
ax_plot_sn(ax, cmip=cmip_stand, cobe=metrics_cobe_stand)

ax = axes.flatten()[2]
ax.set_title('UnforcedAnomaly', fontsize=12)
ax_plot_sn(ax, cmip=cmip_unforced, cobe=metrics_cobe_unforced)

ax = axes.flatten()[3]
ax.set_title('Month Anomaly', fontsize=12)
ax_plot_sn(ax, cmip=cmip_month, cobe=metrics_cobe_month)

ax = axes.flatten()[4]
ax.set_title('Month StdAnomaly', fontsize=12)
ax_plot_sn(ax, cmip=cmip_month_stand, cobe=metrics_cobe_month_stand)

ax = axes.flatten()[5]
ax.set_title('Month UnforcedAnomaly', fontsize=12)
ax_plot_sn(ax, cmip=cmip_month_unforced, cobe=metrics_cobe_month_unforced, cmip_label=True)

ax.set_xticks([-2, 0, 2, 4, 6, 8, 10, 12])
hs, ls = ax.get_legend_handles_labels()
ls.pop(-1)
hs.pop(-1)

fig.legend(hs, ls, bbox_to_anchor=(.85, 1.1), bbox_transform=fig.transFigure, ncol=4, fontsize=12)
fig.supylabel('Probability Density')
fig.supxlabel('S/N Ratio')
plt.tight_layout()
plt.savefig('SNratio-tos.png', dpi=160, bbox_inches='tight')
```
