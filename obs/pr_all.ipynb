```python
def plot_ax(ax, metrics, color='tab:red', label=None):
    upper = np.array(metrics['s_obs_list'])+np.array(metrics['obs_se'])
    lower = np.array(metrics['s_obs_list'])-np.array(metrics['obs_se'])
    center = np.array(metrics['s_obs_list'])
    # ax.plot(upper/np.array(metrics['n_list']), color=color, alpha=.5)
    # ax.plot(lower/np.array(metrics['n_list']), color=color, alpha=.5)
    if label is None:
        ax.plot(metrics['sn_obs'], color=color)
    else:
        ax.plot(metrics['sn_obs'], color=color, label=label)
    ax.fill_between(x=np.arange(444), y1=upper/np.array(metrics['n_list']), y2=lower/np.array(metrics['n_list']), 
                    color=color, alpha=.2)
    ax.plot(metrics['sn'])
    ax.plot(np.arange(center.shape[0]), np.ones_like(center)*2, color='black', linestyle='--')
```

```python
def get_detection_obs(metrics):
    upper = np.array(metrics['s_obs_list'])+np.array(metrics['obs_se'])
    lower = np.array(metrics['s_obs_list'])-np.array(metrics['obs_se'])
    center = np.array(metrics['s_obs_list'])
    upper = upper/np.array(metrics['n_list'])
    lower = lower/np.array(metrics['n_list'])
    center = center/np.array(metrics['n_list'])
    upper = xr.DataArray(upper, dims=['time'], coords={'time': time})
    lower = xr.DataArray(lower, dims=['time'], coords={'time': time})
    center = xr.DataArray(center, dims=['time'], coords={'time': time})

    upper = upper.reindex(time=list(reversed(upper.time)))
    lower = lower.reindex(time=list(reversed(lower.time)))
    center = center.reindex(time=list(reversed(center.time)))

    indices = (upper < 2)
    first_indices = (indices.argmax(dim='time'))
    whole_period = upper.time.shape[0]
    back = whole_period-first_indices
    if back<whole_period:
        detect_time_upper = back
    else:
        detect_time_upper = None
    
    indices = (lower < 2)
    first_indices = (indices.argmax(dim='time'))
    whole_period = lower.time.shape[0]
    back = whole_period-first_indices
    if back<whole_period:
        detect_time_lower = back
    else:
        detect_time_lower = None

    indices = (center < 2)
    first_indices = (indices.argmax(dim='time'))
    whole_period = center.time.shape[0]
    back = whole_period-first_indices
    if back<whole_period:
        detect_time_center = back
    else:
        detect_time_center = None
    return detect_time_lower, detect_time_center, detect_time_upper
```

```python
def plot_detection_ax(ax, detect, gpcp, mswep, cmip_label=False):
    start_month = np.datetime64('1984-01')
    xmin = 40
    xmax = 444
    member = len(detect)
    label_flag = 1
    for i, back in enumerate(detect):
        # plt.hist(back, alpha=0.2, density=True, color=colors[i])
        member = len(back)
        # print(member)
        if member>10:
            mu, std = norm.fit(back)
            print(i, mu, std)
            x = np.linspace(xmin, xmax, 100)
            p = norm.pdf(x, mu, std)
            if cmip_label:
                ax.plot(x, p, linewidth=1, color=colors[i], label=models[i])
            else:
                ax.plot(x, p, linewidth=1, color=colors[i])
    detect_time_lower, detect_time_center, detect_time_upper = get_detection_obs(metrics=gpcp)
    if (detect_time_lower is not None) and (detect_time_center is not None) and (detect_time_upper is not None):
        if cmip_label:
            ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=gpcp_color, linewidth=2, label='GPCP')
        else:
            ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=gpcp_color, linewidth=2)
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=gpcp_color)
        print(detect_time_center.data)
        print('GPCP ', start_month+np.timedelta64(int(detect_time_center.data),'M'))
    if detect_time_lower is None:
        detect_time_lower = xmax
        uncertain = None
    else:
        uncertain = detect_time_lower.data-detect_time_upper.data
    if uncertain is None:
        # ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=gpcp_color)
        label_flag = label_flag-0.5
        pass
    else:
        d_month = start_month+np.timedelta64(int(detect_time_center.data),'M')
        d_month = pd.to_datetime(str(d_month)) 
        d_center = d_month.strftime('%Y %b')
        d_month = start_month+np.timedelta64(int(detect_time_lower.data),'M')
        d_month = pd.to_datetime(str(d_month)) 
        d_upper = d_month.strftime('%Y %b')
        d_month = start_month+np.timedelta64(int(detect_time_upper.data),'M')
        d_month = pd.to_datetime(str(d_month)) 
        d_lower = d_month.strftime('%Y %b')
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=gpcp_color, label='GPCP '+d_center+'\n('+d_lower+'-'+d_upper+')')

    detect_time_lower, detect_time_center, detect_time_upper = get_detection_obs(metrics=mswep)
    if (detect_time_lower is not None) and (detect_time_center is not None) and (detect_time_upper is not None):
        if cmip_label:
            ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=mswep_color, linewidth=2, label='MSWEP')
        else:
            ax.axvline(x=detect_time_center, ymin=0, ymax=1, color=mswep_color, linewidth=2)
        print('MSWEP ', start_month+np.timedelta64(int(detect_time_center.data),'M'))
    if detect_time_lower is None:
        detect_time_lower = xmax
        uncertain = None
    else:
        uncertain = detect_time_lower.data-detect_time_upper.data
    if uncertain is None:
        # ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=mswep_color)
        label_flag = label_flag-0.5
        pass
    else:
        d_month = start_month+np.timedelta64(int(detect_time_center.data),'M')
        d_month = pd.to_datetime(str(d_month)) 
        d_center = d_month.strftime('%Y %b')
        d_month = start_month+np.timedelta64(int(detect_time_lower.data),'M')
        d_month = pd.to_datetime(str(d_month)) 
        d_upper = d_month.strftime('%Y %b')
        d_month = start_month+np.timedelta64(int(detect_time_upper.data),'M')
        d_month = pd.to_datetime(str(d_month)) 
        d_lower = d_month.strftime('%Y %b')
        ax.axvspan(detect_time_lower, detect_time_upper, alpha=0.2, color=mswep_color, label='MSWEP '+d_center+'\n('+d_lower+'-'+d_upper+')')
    if cmip_label:
        hs, ls = ax.get_legend_handles_labels()
        hs = [hs[-3], hs[-1]]
        ls = [ls[-3], ls[-1]]
        ax.legend(hs, ls, loc='upper right')
    else:
        if label_flag>0:
            ax.legend(loc='upper right')
```

```python
def ax_plot_sn(ax, cmip, gpcp, mswep, cmip_label=False, xmax=10):
    for i in range(5):
        data = cmip['model_results'][i]
        data = data['sn'].isel(time=-1)
        mu, std = norm.fit(data)
        x = np.linspace(xmin, xmax, 100)
        p = norm.pdf(x, mu, std)
        if cmip_label:
            ax.plot(x, p, linewidth=1, color=colors[i], label=models[i])
        else:
            ax.plot(x, p, linewidth=1, color=colors[i])
    s_obs_list = gpcp['s_obs_list']
    s_se = gpcp['obs_se']
    n_list = gpcp['n_list']
    center = s_obs_list[-1]/n_list[-1]
    upper = (s_obs_list[-1]+s_se[-1])/n_list[-1]
    lower = (s_obs_list[-1]-s_se[-1])/n_list[-1]
    uncertain = upper.data-lower.data
    if cmip_label:
        ax.axvline(x=center, ymin=0, ymax=1, color=gpcp_color, linewidth=2, label='GPCP')
    else:
        ax.axvline(x=center, ymin=0, ymax=1, color=gpcp_color, linewidth=2)
    ax.axvspan(upper, lower, alpha=0.2, color=gpcp_color, label='GPCP '+'{:.2f}'.format(center)+r'$\pm$'+str(round(uncertain, 2)))
    print(center)
    s_obs_list = mswep['s_obs_list']
    s_se = mswep['obs_se']
    n_list = mswep['n_list']
    center = s_obs_list[-1]/n_list[-1]
    upper = (s_obs_list[-1]+s_se[-1])/n_list[-1]
    lower = (s_obs_list[-1]-s_se[-1])/n_list[-1]
    uncertain = upper.data-lower.data
    if cmip_label:
        ax.axvline(x=center, ymin=0, ymax=1, color=mswep_color, linewidth=2, label='MSWEP')
    else:
        ax.axvline(x=center, ymin=0, ymax=1, color=mswep_color, linewidth=2)
    ax.axvspan(upper, lower, alpha=0.2, color=mswep_color, label='MSWEP '+'{:.2f}'.format(center)+r'$\pm$'+str(round(uncertain, 2)))
    ax.axvline(x=2, ymin=0, ymax=1, color='black', linewidth=1, linestyle='--')
    print(center)
    
    if cmip_label:
        hs, ls = ax.get_legend_handles_labels()
        hs = [hs[-3], hs[-1]]
        ls = [ls[-3], ls[-1]]
        ax.legend(hs, ls, loc='upper right')
    else:
        ax.legend(loc='upper right')
    ax.set_xticks([-2, 0, 2, 4, 6, 8, 10])
```
